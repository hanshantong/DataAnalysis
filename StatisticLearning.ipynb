均值，中位数，众数，极差，

总体方差：$\sigma^2=\frac{1}{N} \displaystyle\sum_{i=1}^{N} (X_i-\mu)^2$<br>
样本方差：$\sigma^2=\frac{1}{N-1} \displaystyle\sum_{i=1}^{N} (X_i-\mu)^2$<br>
将上式展开，可以简化得到：<br>
$\sigma^2 = \frac{1}{N} \displaystyle\sum_{i=1}^{N} X_{i}^2 - \mu^2$ <br>
举例：

```python
#计算方差
def cal_var(data):
    if data is None:
        return
    try:
        m = np.mean(data)
        var = 1.0/len(data) * np.sum(data**2) - m**2 
        return var
    except:
        pass
```

```python
a1 = np.array([1, 2, 5, 8, 9])
b1 = np.array([3, 4, 5, 6, 7])
y1 = [1]*len(a1)
y2 = [1.002]*len(b1)
```

```python
fig, axes = plt.subplots()
axes.scatter(a1, y1, color='r')
axes.scatter(b1, y2, color='b')
plt.legend(['a1', 'b1'])
```

```python
a1_var = a1.var() #std()是标准差
b1_var = b1.var() #std()是标准差
print('a1_var:', a1_var)
print('b1_var:', b1_var)
```

通过a1和b1的方差可以看出，b1的反差比较小，从a1和b1的分布图形来看，显然b1的数据更加集中，对应的方差越小，a1的数据分散，对应的方差较大。因此方差可以衡量数据程度的一个统计量。

>有时候方差会比数据的所有值都大，此时对方差做标准化，即对方差开方。

```python
print(cal_var(a1))
```

```python
a1_sqrt = np.sqrt(a1_var)
b1_sqrt = np.sqrt(b1_var)
print("a1_sqrt:", a1_sqrt)
print("b1_sqrt:", b1_sqrt)
```

### 6.线性代数

#### 6.1范数

范数是将向量映射到非负值的函数，范数是满足下列性质的任意函数：

（1）$f(x)=0 \implies x=0$  
（2）$f(x+y) \ll f(x) + f(y)$  
（3）$\forall \alpha \in R, f(\alpha x) = |\alpha|f(x)$

矩阵A的F范数:$\|A\|_F = \sqrt{\displaystyle\sum_{i,j}A_{i,j}^2}$ <br>
计算如下：

```python
A = np.random.randint(0, 120, size=(20, 20))
ans = np.sqrt(np.sum(A**2))
print(ans)
```

向量$\textbf{x}$的$\textit{P}$范数$L^P$定义如下：  
       $\quad \quad \quad \quad \quad \quad \quad \quad $ $\|x\|_P = (\displaystyle\sum_{i}|x_i|^P)^{\frac{1}{P}}$

>当P等于2时，$L^2$叫做欧几里得范数。

```python
B = np.triu(A)
```

```python
B = B + B.T - np.diag(B.diagonal())
```

```python
np.allclose(B, B.T)
```

#### 6.1特征分解

假设矩阵$\textbf{A}$具有n个线性无关的特征向量$\{v^{(1)}, v^{(2)}, v^{(3)},..., v^{(n)}\}$，对应的特征值是$\{\lambda_1, \lambda_2, \lambda_3, ..., \lambda_n\}$，将特征向量组成一个矩阵，使得每一列对应一个特征向量，即$V = [v^{(1)}, v^{(2)}, v^{(3)},..., v^{(n)}]$,对应地$\lambda = [\lambda_1, \lambda_2, \lambda_3, ..., \lambda_n]$,那么矩阵$\textbf{A}$的特征分解是：  
$\quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad A = Vdiag(\lambda)V^{-1}$

举例如下：

```python
val, vec = np.linalg.eig(B) #求矩阵B的特征值和特征向量
```

```python
C = vec @ np.diag(val) @ np.linalg.inv(vec)
```

验证：

```python
np.allclose(B, C)
```

#### 6.2奇异值分解

对于非方阵$\textbf{A}$，可以将其分解为三个矩阵的乘积：  
$ \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad A = UDV^T$  
如果$\textbf{A}$是一个$m \times n$的矩阵，那么$U$是$m \times m$的矩阵，$D$是$m \times n$的矩阵，$V$是$n \times n$的矩阵。

```python
A = np.random.randint(1,100, size=(5, 4))
```

```python
np.linalg.svd?
```

```python
u, s, vh = np.linalg.svd(A, full_matrices=False)
```

#### 6.3伪逆

矩阵（非方阵）$\textbf{A}$的伪逆定义如下：  
$A^+ = \lim\limits_{\alpha \to 0}(A^TA + \alpha I)^{-1} A^T$  
但是通常在计算$\textbf{A}$的伪逆时，不是使用上述的公式，而是使用如下的公式进行计算：  
$\quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad A^+ = VD^+U^T$  
其中，$U,D,V$是矩阵$\textbf{A}$的奇异值分解，对角矩阵$\textbf{D}$的伪逆$D^+$是其非零元素取倒数后转置得到的。

```python
A_p = np.linalg.pinv(A) #使用numpy.linalg计算A的伪逆
```

```python
A_svd = vh.T @ np.diag(1/s).T @ u.T #采用公式计算A的伪逆
```

```python
np.allclose(A_p, A_svd) #比较伪逆是否一致
```
