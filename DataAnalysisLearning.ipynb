### 布尔索引

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
```

```python
data = np.random.randn(7, 4)
```

```python
names
```

```python
data
```

假设data中的一行数据对应names中的一个人。注意：同一个人可以对应数据，那么可以通过布尔索引取出某个人所对应的的数据，如下所示：

```python
names == 'Bob'  #返回一个布尔数组
```

```python
#使用上面产生的布尔数组作为索引去取出Bob对应的行数据
data[names == 'Bob']
```

可以看到，布尔数组中为True的索引处数据被取了出来。在上述的例子是第0行和第3行数据。

>布尔数组作为索引时其长度必须和所要索引的轴的长度一致，虽然不这样做不会导致异常，但是可能会得到意想不到的结果，所以使用这个特性时要多加注意。

```python
data[names=='Bob', 2:] #去除Bob的行数据，并只取第2列往后的数据
```

```python
data[names=='Bob', 3] #取出Bob的行数据，并只取第3列
```

```python
#to select everything but "Bob", you can either use != or 
#negate the condiction using ~
data[names != 'Bob']
```

```python
data[~(names=='Bob')]
```

波浪号(~)用于对一个布尔条件进行取反

```python
cond = names == 'Bob'
```

```python
data[~cond]
```

如果选择多个人对应的数据，可以使用逻辑运算符如“&”和“|”

```python
mask = (names == 'Bob') | (names=='Will')
```

```python
mask
```

```python
data[mask]
```

>注意：通过布尔索引取出的数据都是深度拷贝的，即使返回的数组没有任何变化。

```python
#将data中所有小于0的数据都置为0,此时该操作会直接反映到原数组中
data[data < 0] = 0
```

### 将条件逻辑表述为数组操作
#### Expressing Conditional Logic as Array Operations

The numpy.where function is a vecterized version of the ternary(三元的) expression x if condition else y. Suppose we have a boolean array and two arrays of values:

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
```

```python
cond = np.array([True, False, True, False, False])
```

Suppose we want to take a value from *xarr* whenever the corresponding value in *cond* is True, and otherwise take the value from *yarr*. A list comprehension doing this mihgt look like:

```python
result = [x if c else y for x, y, c in zip(xarr, yarr, cond)]
```

```python
result
```

上述做饭会有很多问题，第一，对于大型数组，操作会很慢；第二，它不适用多维数组。numpy.where解决上述的问题：

```python
result = np.where(cond, xarr, yarr)
```

```python
result
```

>numpy.where()的第二和第三个参数并不一定是数组，他们两者任一个或者两者均可以是普通的常数。

假设我们想将一个数组中小于0的所有数据都替换为-5，所有大于0的替换为5，可以使用numpy.where()实现：

```python
arr = np.random.randn(5,5)
```

```python
arr
```

```python
arr > 0
```

```python
np.where(arr > 0, 5, -5)
```

```python
#将arr中大于0的替换为2，其它值保持不变
np.where(arr > 0, 2, arr)
```

```python
bools.all() # 是否所有值都是True
```

### 例子：随机游走
#### Example： Random Walk

```python
import random
```

```python
position = 0
walk = [position]
steps = 1000 # 走1000步
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk[:100])
```

```python
nsteps = 1000
draws = np.random.randint(0, 2, nsteps)
```

```python
steps = np.where(draws > 0, 1, -1)
```

```python
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
arr = np.random.randint(0, 100, size=8)
print(arr)
```

```python
arr.argmax()
```

A more comlicated statistic is the *first crossing time (首次穿越时间)*, the step at which the random walk reaches a particular values. Here we might want to know how long it took the random walk to get at least 10 steps away from the origin 0 in either direction. np.abs(walk) >= 10 gives us a boolean array indicatiing where the walk has reached or exceeded 10, but we want the index of the first 10 or -10. Turns out, we can compute this using *argmax*, which returns the first index of the maximum valueint the boolean array (True is the maximum value):

```python
np.abs(walk >= 10).argmax()
```
