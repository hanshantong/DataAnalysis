```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import pi
```

```python
x = np.linspace(0, 2 * pi,360)
```

```python
y = np.cos(np.exp(x))+x
```

```python
plt.plot(x, y)
plt.grid()
plt.xlabel('x轴:时间', fontproperties="SimHei", fontsize=16, color='red')
plt.ylabel("y轴：振幅", fontproperties="SimHei", fontsize=18, color='cyan')
plt.title('$y=cos(e^{x}) + x$')
plt.show()
```
