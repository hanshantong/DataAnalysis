```python
def find_all_indixes(a, val):
    '''
    find all the indexes of val in a
    
    parameters：
    a：an 1D array-like.
    val: the values that is required to find all the indexes.
    
    returns:
    a list of
    '''
    ans = [i for i, value in enumerate(a) if value == val]
    #or
    #ans=list(filter(lambda x:a[x] == val, range(len(a))))
    return ans
```

```python
def get_frequency(num, start_freq, stop_freq, points_freq):
    '''
    获取第num个频点对应的频率值，计算公式如下：
    freq = start_freq + span_freq / (point_freq - 1) * (num - 1)
    span_freq = stop_freq - start_freq
    center_Freq = (start_freq + stop_freq) / 2.0
    
    参数：
    num：int，非负数（non-negative）
         表示要获取第num个频点对应的频率。
    start_freq：float
         表示扫描的起始频率。
    stop_freq：float
         表示扫描的停止频率。
    points_freq：int，非负数（non-negative）
         表示扫描的频点个数
         
    返回值：
    一个float值，表示第num个频点对应的频率。
    '''
    if points_freq == 1:
        return start_freq
    if num > points_freq:
        raise ValueError('num:' + str(num) + " must be less equal than " + 'points_freq:' + str(points_freq))
    if start_freq > stop_freq:
        raise ValueError("start_freq:" + str(start_freq) + " must be less equal than stop_freq:" + str(stop_freq))
        
    span_freq = stop_freq - start_freq
    freq = start_freq + span_freq / (points_freq - 1) * (num - 1)
    return freq
```

```python
def approximate_size(size):
    '''
    将文件大小显示人可以方便查看的形式
    Arguments:
    size -- the size of the file in bytes
    a_kilobyte_is_1024_bytes -- indicates whether use 1024 or 1000 to compute file size

    Returns:
    a string 

    ''' 
    
    SUFFIXES = {1000:['KHz', 'MHz', 'GHz', 'THz']}
    if size < 0:
        raise ValueError("size must be non-negative!")
    multiples = 1000
    for suffix in SUFFIXES[multiples]:
        size /= multiples
        if size < multiples:
            return "{0:.5f}{1}".format(size, suffix)
    raise ValueError('size is too large.')
```
